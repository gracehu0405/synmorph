```python
import numpy as np
import sys
import os
sys.path.append(os.path.abspath('..'))
import synmorph as sm

# change to match our defaults
tissue_params = {"L": 9,                   # domain size(?)
                 "A0": 1,                  # reference/initial area
                 "P0": 3.81,               # reference/initial perimeter
                 "kappa_A": 1,             # area modulus (resistance to changing cell area)
                 "kappa_P": 0.1,           # perimeter modulus
                 "W": np.array(((0, 0.0762), (0.0762, 0)))*0.75, # cell-cell interaction matrix, 0.75 scaling factor, + values mean repulsion
                 "a": 0,                   # maximal radius to engage in repulsion
                 "k": 0}                   # coefficient of repulsion
active_params = {"v0": 1e-1,               # Instantaneous velocity -- "activity" for L, M, ECM cells
                 "Dr": 1e-1}               # cell persistence for how much a cell keeps in the same direction
init_params = {"init_noise": 0.005,        # initial arrangement noise 0.01
               "c_type_proportions": (0.5, 0.5)} # 50% red, 50% blue cells
run_options = {"equiangulate": True,       # ?
               "equi_nkill": 3}            # ?
simulation_params = {"dt": 0.025,          # simulation time step
                     "tfin": 250,          # simulation time end point
                     "tskip": 10,          # saving every 10th step in the movie
                     "grn_sim": None}      # what does grn stand for ? perhaps signaling component
save_options = {"save": "last",            # minimal saving option, as an alternative. Saves the simulation class instantiation to a pickle file
                "result_dir": "results",   # save results to results directory
                "name": "ctype_example_2", # file name prefix to be saved
                "compressed": True}        # bzip2 compression
sim = sm.simulation(tissue_params=tissue_params,
                    active_params=active_params,
                    init_params=init_params,
                    simulation_params=simulation_params,
                    run_options=run_options,
                    save_options=save_options)
sim.simulate(progress_bar=True)

sim.animate_c_types(n_frames=100,
                    c_type_col_map=["#D67D79", "#686EA1"],
                    file_name=None)

# Public repo synmorph ok?
```
